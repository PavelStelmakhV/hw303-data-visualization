```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
stock_return = pd.DataFrame(
    {
        "Доходность A, %": [25, -10, 10, 5, 35, 13],
        "Доходность B, %": [0, 15, -5, 5, 20, 25],
        "Доходность C, %": [10, 25, -15, -5, -5, 15],
    },
    index=[1, 2, 3, 4, 5, 6],
)
```

```python
%matplotlib inline
sns.regplot(data=stock_return, x="Доходность A, %", y="Доходность B, %")
plt.show()
```

```python
%matplotlib inline
sns.regplot(data=stock_return, x="Доходность A, %", y="Доходность C, %", color='green')
plt.show()
```

```python
%matplotlib inline
sns.regplot(data=stock_return, x="Доходность B, %", y="Доходность C, %", color='orange')
plt.show()
```
