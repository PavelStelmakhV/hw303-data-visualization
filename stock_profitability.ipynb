```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
stock_return = pd.DataFrame(
    {
        "Доходность A": [0.25, -0.1, 0.1, 0.05, 0.35, 0.13],
        "Доходность B": [0, 0.15, -0.05, 0.05, 0.2, 0.25],
        "Доходность C": [0.1, 0.25, -0.15, -0.05, -0.05, 0.15],
    },
    index=[1, 2, 3, 4, 5, 6],
)
```

```python
%matplotlib inline
sns.regplot(data=stock_return, x="Доходность A", y="Доходность B")
plt.show()
```

```python
%matplotlib inline
sns.regplot(data=stock_return, x="Доходность A", y="Доходность C", color='green')
plt.show()
```

```python
%matplotlib inline
sns.regplot(data=stock_return, x="Доходность B", y="Доходность C", color='orange')
plt.show()
```
